```python
def fib(n):
    if n <=1:
        return n
    return fib(n-1)+fib(n-2)


print(fib(2))
print(fib(10))
print(fib(30))
print(fib(40)) # runs 60 seconds
print(fib(50)) # runs a long time
```

```python
def fibDP(n):
    memo = {}
    return fibMem(n,memo)

def fibMem(n, memo):
    if n in memo:
        return memo[n]
    if n <= 1:
        memo[n] = n
    else:
        memo[n] = fibMem(n-1,memo)+fibMem(n-2,memo)
    return memo[n]


print(fibDP(2))
print(fibDP(10))
print(fibDP(50))
print(fibDP(800))
print(fibDP(1000)) # throws recursion error
# print(fibDP(100000)) # throws recursion error
```

```python
def fibDPBU(n):
    memo = {}
    memo[0] = 0
    memo[1] = 1
    for i in range(2,n+1):
        memo[i] = memo[i-1]+memo[i-2]
    return memo[n]
    
print(fibDPBU(2))
print(fibDPBU(10))
print(fibDPBU(50))
print(fibDPBU(800))
print(fibDPBU(1000))
print(fibDPBU(100000))
```

```python
import time
# Splitting into coins recursively (inefficient)
# depends on the fact that last coin is 1

coin = [200, 100, 50, 20, 10, 5, 2, 1]

# add a counter so that we can check how many recursive calls are made
c1 = 0

def coinSplit(m):
    return coinSplitRec(m,0)

def coinSplitRec(m, lo):
    global c1
    c1 += 1
    if m == 0:
        return 0
    if lo == len(coin)-1:
        return m
    withoutIt = coinSplitRec(m,lo+1)
    if coin[lo] <= m:
        withIt = 1 + coinSplitRec(m-coin[lo],lo)
        if withIt < withoutIt:
            return withIt
    return withoutIt

# add a counter so that we can check how many recursive calls are made
c2 = 0

def coinSplitDP(m):
    memo = {}
    return coinSplitMem(m,0,memo)

def coinSplitMem(m, lo, memo):
    global c2
    c2 += 1    
    if (m,lo) in memo:
        return memo[m,lo]
    if m == 0:
        memo[m,lo] = 0
    elif lo == len(coin)-1:
        memo[m,lo] = m
    else:
        withoutIt = coinSplitMem(m,lo+1,memo)
        memo[m,lo] = withoutIt
        if coin[lo] <= m:
            withIt = 1 + coinSplitMem(m-coin[lo],lo,memo)
            if withIt < withoutIt:
                memo[m,lo] = withIt
    return memo[m,lo]

def coinSplitDPBU(mInit):
    memo = {}
    for i in range(len(coin)):
        memo[0,i] = 0
    for m in range(mInit+1):
        memo[m,len(coin)-1] = m
    for m in range(1,mInit+1):
        for lo in range(len(coin)-2,-1,-1):
            withoutIt = memo[m,lo+1]
            memo[m,lo] = withoutIt
            if coin[lo] <= m:
                withIt = 1 + memo[m-coin[lo],lo]
                if withIt < withoutIt:
                    memo[m,lo] = withIt
    return memo[mInit,0]

# print(coinSplit(599), coinSplitDP(599), coinSplitDPBU(599), c1, c2)

# coin = [i for i in range(200,0,-1)]
# coinSplitDPBU(100)

coin1 = [200, 100, 50, 20, 5, 2, 1] 
coin2 = [1 for i in range(200)]
coin2 = coin2[::-1]

#question 3
def coinSplitTime(n):
    t = time.time()
    for i in range(len(coin2)):
        coinSplit(n/coin2[i])
    t = time.time() - t
    return t

#testing coinSplitDPTime(n)
#print(coinSplitTime(10000))
    
def coinSplitDPTime(n):
    t = time.time()
    for i in range(len(coin1)):
        coinSplitDP(n/coin1[i])
    t = time.time() - t
    return t

#testing coinSplitDPTime(n)
#print(coinSplitDPTime(100))
    
def coinSplitDPBUTime(n):
    t = time.time()
    for i in range(len(coin2)):
        coinSplitDPBU(n//coin2[i])
    t = time.time()-t
    return t

#testing coinsplitDPBUTime(n)
print(coinSplitDPBUTime(10000))
#print("hello")
```

```python
def longestPalinDP(s):
    if len(s) == 0:
        return s
    memo = {}
    return palinMem(s,0,len(s),memo)

def palinMem(s, lo, hi, memo):
    if (lo,hi) in memo:
        return memo[lo,hi]
    if hi-lo <= 1:
        memo[lo,hi] = s[lo:hi]
    else:
        if s[lo] == s[hi-1]:
            memo[lo,hi] = s[lo]+palinMem(s,lo+1,hi-1,memo)+s[hi-1]
        else:
            s1 = palinMem(s,lo,hi-1,memo)
            s2 = palinMem(s,lo+1,hi,memo)
            if len(s1) < len(s2):
                memo[lo,hi] = s2
            else:
                memo[lo,hi] = s1
    return memo[lo,hi]

print(longestPalinDP(""))
print(longestPalinDP("10"))
print(longestPalinDP("010"))
print(longestPalinDP("10001"))
print(longestPalinDP("12312323321"))
print(longestPalinDP("sdwqeqwewarasdaeaweawrafd"))
print(longestPalinDP("sdwqeqwewarasdaeaweawrafdsdwqeqwewarasdaeaweawrafd"))
```
